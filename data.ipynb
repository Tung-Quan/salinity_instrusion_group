```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# ======== LOAD TRAINING DATA ========
train_dfs = []
for year_file in ['MAN 2022.xlsx', 'MAN 2023.xlsx']:
    sheets = pd.read_excel(year_file, sheet_name=None, skiprows=4, nrows=31, na_filter=True)
    for sheet_name, df_sheet in sheets.items():
        df_sheet = df_sheet.dropna(how='all')
        if df_sheet.shape[1] < 3:
            continue
        df_selected = df_sheet.iloc[:, 0:3].dropna().copy()
        df_selected[df_selected.columns[1]] = pd.to_numeric(df_selected[df_selected.columns[1]], errors='coerce')
        df_selected[df_selected.columns[2]] = pd.to_numeric(df_selected[df_selected.columns[2]], errors='coerce')
        df_selected = df_selected.dropna()
        train_dfs.append(df_selected)

train_data = pd.concat(train_dfs, ignore_index=True)

# ======== LOAD TEST DATA ========
test_dfs = []
sheets_test = pd.read_excel('MAN 2024.xlsx', sheet_name=None, skiprows=4, nrows=31, na_filter=True)
for sheet_name, df_sheet in sheets_test.items():
    df_sheet = df_sheet.dropna(how='all')
    if df_sheet.shape[1] < 3:
        continue
    df_selected = df_sheet.iloc[:, 0:3].dropna().copy()
    df_selected[df_selected.columns[1]] = pd.to_numeric(df_selected[df_selected.columns[1]], errors='coerce')
    df_selected[df_selected.columns[2]] = pd.to_numeric(df_selected[df_selected.columns[2]], errors='coerce')
    df_selected = df_selected.dropna()
    test_dfs.append(df_selected)

test_data = pd.concat(test_dfs, ignore_index=True)

# ======== FEATURE AND TARGET =========
feature_col = train_data.columns[1]
target_col = train_data.columns[2]

X_train = train_data[[feature_col]]
y_train = (train_data[target_col] > train_data[target_col].median()).astype(int)

X_test = test_data[[feature_col]]
y_test = (test_data[target_col] > test_data[target_col].median()).astype(int)

# ======== SCALING (TRNH DATA LEAKAGE) =========
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ======== MODEL EVALUATION =========
models = {
    "LogisticRegression": LogisticRegression(class_weight='balanced', random_state=42),
    "RandomForest": RandomForestClassifier(class_weight='balanced', random_state=42),
    "SVM": SVC(class_weight='balanced', probability=True, random_state=42)
}

for name, clf in models.items():
    clf.fit(X_train_scaled, y_train)
    y_pred = clf.predict(X_test_scaled)
    print(f"\n Model: {name}")
    print(classification_report(y_test, y_pred))

# ======== OPTIONAL: TUNING LOGISTIC REGRESSION ========
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100]
}
grid = GridSearchCV(LogisticRegression(class_weight='balanced', random_state=42), 
                    param_grid, cv=5, scoring='f1')
grid.fit(X_train_scaled, y_train)
best_model = grid.best_estimator_
y_pred = best_model.predict(X_test_scaled)
print("\n===Best LogisticRegression (after GridSearchCV):")
print(classification_report(y_test, y_pred))

# ======== PLOTTING DISTRIBUTIONS ========
sns.histplot(train_data[target_col], kde=True)
plt.title("Target Distribution in Training Set")
plt.xlabel("Target Values")
plt.show()

sns.boxplot(data=train_data, x=y_train, y=feature_col)
plt.title("Feature vs Target")
plt.xlabel("Binary Label")
plt.ylabel("Feature Value")
plt.show()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# ======= Hm to chui sliding window nhiu ct =======
def create_time_series_multicol(df, feature_cols, target_cols, window_size=3):
    X, y = [], []
    for i in range(len(df) - window_size):
        window_features = df[feature_cols].iloc[i:i+window_size].values.flatten()
        next_targets = df[target_cols].iloc[i + window_size].values
        X.append(window_features)
        y.append(next_targets)
    return pd.DataFrame(X), pd.DataFrame(y, columns=target_cols)

# ======= Load data nm 2022 + 2023 =======
train_dfs = []
for year_file in ['MAN 2022.xlsx', 'MAN 2023.xlsx']:
    sheets = pd.read_excel(year_file, sheet_name=None, skiprows=4, nrows=31)
    for df in sheets.values():
        df = df.dropna(how='all')
        if df.shape[1] < 3:
            continue
        df = df.iloc[:, 0:3].dropna()
        df.columns = ['time', 'smax', 'smin']
        df['smax'] = pd.to_numeric(df['smax'], errors='coerce')
        df['smin'] = pd.to_numeric(df['smin'], errors='coerce')
        df = df.dropna()
        train_dfs.append(df)

train_data = pd.concat(train_dfs, ignore_index=True)

# ======= Chun ha d liu =======
scaler = StandardScaler()
train_data[['smax', 'smin']] = scaler.fit_transform(train_data[['smax', 'smin']])

# ======= To sliding window train =======
X_train, y_train = create_time_series_multicol(train_data, ['smax', 'smin'], ['smax', 'smin'], window_size=3)

# ======= Load test data nm 2024 =======
test_dfs = []
sheets = pd.read_excel('MAN 2024.xlsx', sheet_name=None, skiprows=4, nrows=31)
for df in sheets.values():
    df = df.dropna(how='all')
    if df.shape[1] < 3:
        continue
    df = df.iloc[:, 0:3].dropna()
    df.columns = ['time', 'smax', 'smin']
    df['smax'] = pd.to_numeric(df['smax'], errors='coerce')
    df['smin'] = pd.to_numeric(df['smin'], errors='coerce')
    df = df.dropna()
    test_dfs.append(df)

test_data = pd.concat(test_dfs, ignore_index=True)
test_data[['smax', 'smin']] = scaler.transform(test_data[['smax', 'smin']])

# ======= To sliding window test =======
X_test, y_test = create_time_series_multicol(test_data, ['smax', 'smin'], ['smax', 'smin'], window_size=3)

# ======= Hun luyn m hnh a u ra =======
base_model = RandomForestRegressor(random_state=42)
model = MultiOutputRegressor(base_model)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# ======= nh gi kt qu =======
for i, target in enumerate(['smax', 'smin']):
    print(f" Regression for {target}")
    print("  MSE:", mean_squared_error(y_test[target], y_pred[:, i]))
    print("  R :", r2_score(y_test[target], y_pred[:, i]))
    print()

# ======= V biu  kt qu =======
for i, target in enumerate(['smax', 'smin']):
    plt.figure()
    plt.plot(y_test[target].values, label='Actual', marker='o')
    plt.plot(y_pred[:, i], label='Predicted', marker='x')
    plt.title(f"D on {target} nm 2024")
    plt.xlabel("Time Step")
    plt.ylabel("Gi tr (scaled)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.multioutput import MultiOutputRegressor
import matplotlib.pyplot as plt

# ======= HM TO CHUI TRAIN =======
def create_time_series_multicol(df, feature_cols, target_cols, window_size=3):
    X, y = [], []
    for i in range(len(df) - window_size):
        window_features = df[feature_cols].iloc[i:i+window_size].values.flatten()
        next_targets = df[target_cols].iloc[i + window_size].values
        X.append(window_features)
        y.append(next_targets)
    return pd.DataFrame(X), pd.DataFrame(y, columns=target_cols)

# ======= Load training data (2022 + 2023) =======
train_dfs = []
for year_file in ['MAN 2022.xlsx', 'MAN 2023.xlsx']:
    sheets = pd.read_excel(year_file, sheet_name=None, skiprows=4, nrows=31)
    for df in sheets.values():
        df = df.dropna(how='all')
        if df.shape[1] < 3:
            continue
        df = df.iloc[:, 0:3].dropna()
        df.columns = ['time', 'smax', 'smin']
        df[['smax', 'smin']] = df[['smax', 'smin']].apply(pd.to_numeric, errors='coerce')
        df = df.dropna()
        train_dfs.append(df)

train_data = pd.concat(train_dfs, ignore_index=True)

# ======= Chun ha d liu =======
scaler = StandardScaler()
train_data[['smax', 'smin']] = scaler.fit_transform(train_data[['smax', 'smin']])

# ======= To tp train =======
X_train, y_train = create_time_series_multicol(train_data, ['smax', 'smin'], ['smax', 'smin'], window_size=3)

# ======= Hun luyn m hnh =======
base_model = RandomForestRegressor(random_state=42)
model = MultiOutputRegressor(base_model)
model.fit(X_train, y_train)

# ======= FORECAST 7 NGY TNG LAI =======
def forecast_future(model, last_days, n_steps, window_size=3):
    forecasts = []
    current_window = last_days.copy()
    for _ in range(n_steps):
        input_seq = current_window[-window_size:].flatten().reshape(1, -1)
        next_pred = model.predict(input_seq)[0]
        forecasts.append(next_pred)
        current_window = np.vstack([current_window, next_pred.reshape(1, -1)])
    return np.array(forecasts)

# ======= Load d liu test (2024)  ly 3 ngy gn nht =======
test_dfs = []
sheets = pd.read_excel('MAN 2024.xlsx', sheet_name=None, skiprows=4, nrows=31)
for df in sheets.values():
    df = df.dropna(how='all')
    if df.shape[1] < 3:
        continue
    df = df.iloc[:, 0:3].dropna()
    df.columns = ['time', 'smax', 'smin']
    df[['smax', 'smin']] = df[['smax', 'smin']].apply(pd.to_numeric, errors='coerce')
    df = df.dropna()
    test_dfs.append(df)

test_data = pd.concat(test_dfs, ignore_index=True)
test_data[['smax', 'smin']] = scaler.transform(test_data[['smax', 'smin']])

# Ly 3 ngy gn nht lm u vo
last_window = test_data[['smax', 'smin']].iloc[-3:].values
future_forecast = forecast_future(model, last_window, n_steps=7)

# ======= V D ON =======
future_forecast_inv = scaler.inverse_transform(future_forecast)

days = [f"Day {i+1}" for i in range(7)]

plt.figure(figsize=(10, 5))
plt.plot(days, future_forecast_inv[:, 0], label="D on smax", marker='o')
plt.plot(days, future_forecast_inv[:, 1], label="D on smin", marker='x')
plt.title("D bo 7 ngy tip theo (smax v smin)")
plt.xlabel("Ngy tng lai")
plt.ylabel("Gi tr gc")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ======= In ra kt qu =======
forecast_df = pd.DataFrame(future_forecast_inv, columns=['smax', 'smin'])
forecast_df.index = days
print(" D bo 7 ngy ti:")
print(forecast_df)
```
